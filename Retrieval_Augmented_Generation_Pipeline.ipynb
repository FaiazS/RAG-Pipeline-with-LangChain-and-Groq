```python
#Loader - To load the Website

from langchain_community.document_loaders import WebBaseLoader

website_loader = WebBaseLoader(

                               web_paths = ["https://www.educosys.com/course/genai"]
)

documents = website_loader.load()

print(documents)
```

```python
#Splitting the documents into smaller chunks

#Overlap of 200 characters to maintain context/meaning/relationship across chunks

from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 200)

splits = text_splitter.split_documents(documents)

print(splits[0])

print(splits[1])

print(splits[2])
```

```python
#Performing Embedding on the Splits and Storing them in the Vector Database(ChromaDB)

from langchain_huggingface import HuggingFaceEmbeddings

from langchain.vectorstores import Chroma

vectorstore = Chroma.from_documents(documents = splits, embedding = HuggingFaceEmbeddings())
```

```python
#Total number of document chunks present

print(vectorstore._collection.count())
```

```python
#Viewing/Displaying the document chunks present

print(vectorstore._collection.get())
```

```python
#Retriever - Responsible for retrieving out the information from the Vector database

information_retriever = vectorstore.as_retriever()
```

```python
#Defining/Pulling a Predefined Retrieval Augmented Generation Prompt Template from LangChain Hub

from langchain import hub

rag_prompts = hub.pull("rlm/rag-prompt")

print(rag_prompts)
```

```python
#Setting up the Large Language Model

import os

from langchain_groq import ChatGroq

from google.colab import userdata

os.environ["GROQ_API_KEY"] = userdata.get("LLMProjectKey1")

llm_model = ChatGroq(model = "llama-3.3-70b-versatile")

llm_response = llm_model.invoke("Hey there I am building my first RAG Pipeline")

print(llm_response)
```

```python
#Formattting/Post-processing the Documents

def format_document(document):

  return "\n\n".join(document.page_content for document in documents)
```

```python
#Defining the Retrieval Augmented Generation Chain

from langchain_core.runnables import RunnablePassthrough

from langchain_core.runnables import RunnableLambda

from langchain_core.output_parsers import StrOutputParser

def print_rag_prompt(rag_prompt):

  print("\n Prompt :", rag_prompt)

  return rag_prompt

rag_chain = ({"context" :  information_retriever | format_document, "question" : RunnablePassthrough()}

             | rag_prompts

             | RunnableLambda(print_rag_prompt)

             | llm_model

             | StrOutputParser())
```

```python
#Testing out the Retrieval Augmentated Generation Pipeline

rag_response1 = rag_chain.invoke("What is Gen AI")

rag_response2 = rag_chain.invoke("Please list out the curriculum for the Gen AI Course")

print("RAG response1 :" , rag_response1)

print("RAG response2 :", rag_response2)
```
